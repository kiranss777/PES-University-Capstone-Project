```python
from skimage.metrics import structural_similarity as ssim
from skimage import io, color, img_as_ubyte
from skimage.transform import resize
import numpy as np
import os
```

```python
def calculate_ssim(image1, image2):
    # Convert images to grayscale if not already
    if image1.shape[-1] == 3:
        image1 = color.rgb2gray(image1)
    if image2.shape[-1] == 3:
        image2 = color.rgb2gray(image2)

    # Ensure images have the same shape
    min_shape = np.min([image1.shape, image2.shape], axis=0)
    image1 = resize(image1, min_shape)
    image2 = resize(image2, min_shape)

    # Ensure images have the same data type
    image1 = img_as_ubyte(image1)
    image2 = img_as_ubyte(image2)

    # Calculate SSIM
    ssim_index, _ = ssim(image1, image2, full=True)
    return ssim_index
```

# **SSIM_PNUEMONIA**

```python
original_folder="/kaggle/input/real-pnuemonia-images/preprocess_VIRAL-PNEUMONIA"
```

```python
image_path1 = "/kaggle/input/best-pnuemonia-progan/best_pnuemonia_progan/generated_image_17_5.png"
image1 = io.imread(image_path1)
```

```python
 
 ssim_values = []

 for original_file in os.listdir(original_folder):
        original_image_path = os.path.join(original_folder, original_file)
        image2= io.imread(original_image_path)
        image1 = resize(image1, (image2.shape[0], image2.shape[1]), anti_aliasing=True)
        ssim_value = calculate_ssim(image1, image2)
        ssim_values.append(ssim_value)
```

```python
print(max(ssim_values))
```
